## Draw a `heart` using Python

### **Heart Curve**

A `heart curve` can be defined parametrically as

$$\begin{eqnarray}
x &=& 16\sin^3(t) \\
y &=& 13\cos(t) &-& 5\cos(2t) &-& 2\cos(3t) &-& \cos(4t) . \\
\end{eqnarray}$$

The above `parameteric equation` was used to plot the `heart` shape. \\
(I refer to the following resource for the parameteric equation: https://mathworld.wolfram.com/HeartCurve.html)

```
import matplotlib.pyplot as plt
import numpy as np
```

Define 3 colors (pink, magenta, red) for the color of hearts.

```
colors = ['pink', 'magenta', 'red']
```

Define the above parameteric equation.

```
t = np.arange(0, 2 * np.pi, 0.1)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t)-np.cos(4 * t)
```

Plot the equation 3 times in different colors (pink, magenta, red).

```
for i in colors:
    plt.plot(x, y, color = i)
    plt.text(-6.65,0, "Professor Hong's", fontsize = 15)
    plt.text(-7,-5, "Machine Learning", fontsize = 15)
    plt.show()
```
